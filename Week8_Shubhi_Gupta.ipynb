# Week 8 Assignment

## Single-Agent AI Pipeline

### Objective
Build a simple AI agent that routes user queries to the correct tool using conditional routing and returns responses in JSON format.

```python
# Import required libraries

import re
import json
```

## Step 1: Calculator Tool

This tool evaluates basic mathematical expressions entered by the user.

```python
def calculator(expression):
    try:
        result = eval(expression)
        return result
    except Exception:
        raise ValueError("Invalid mathematical expression.")
```

## Step 2: Test Calculator

```python
print(calculator("10+5"))
print(calculator("(20*5)/2"))
```

## Step 3: Keyword Extraction Tool

This tool extracts meaningful words by removing common stop words.

```python
def extract_keywords(text):

    stop_words = {
        "the","is","a","an","of","to","and",
        "in","on","for","with","this",
        "that","it","are"
    }

    words = text.lower().split()

    keywords = []

    for word in words:

        word = word.strip(".,!?")

        if word not in stop_words:
            keywords.append(word)

    return keywords
```

## Step 4: Test Keyword Tool

```python
sentence = "Artificial Intelligence is changing the world."

print(extract_keywords(sentence))
```

## Step 5: Build the Agent

The agent decides which tool to use based on the user's request.

```python
def agent(query):

    query = query.strip()

    try:

        lower_query = query.lower()

        if "calculate" in lower_query:

            expression = lower_query.replace("calculate","").strip()

            answer = calculator(expression)

            return {
                "type":"calculation",
                "result":answer
            }

        elif "keywords" in lower_query:

            text = lower_query.replace("keywords","").strip()

            answer = extract_keywords(text)

            return {
                "type":"keywords",
                "result":answer
            }

        else:

            return {
                "type":"general",
                "result":"Sorry, I can only perform calculations and keyword extraction."
            }

    except Exception as e:

        return {
            "type":"error",
            "result":str(e)
        }
```

## Step 6: Test the Agent

```python
print(json.dumps(agent("calculate 10+20"), indent=4))

print(json.dumps(agent("keywords Artificial Intelligence is changing healthcare"), indent=4))

print(json.dumps(agent("Hello"), indent=4))

print(json.dumps(agent("calculate 10/0"), indent=4))
```

## Step 7: Validation Tests

Run multiple queries to verify the routing logic.

```python
test_queries = [

    "calculate 100+25",

    "calculate (50*2)/5",

    "keywords Machine Learning is transforming education",

    "Hello",

    "calculate 5//"

]

for query in test_queries:

    print("Query:", query)

    print(json.dumps(agent(query), indent=4))

    print("-"*40)
```

## Step 8: Interactive Mode

Users can enter multiple queries until they type **exit**.

```python
while True:

    user_query = input("Enter your query (type 'exit' to stop): ")

    if user_query.lower() == "exit":
        print("Agent Closed.")
        break

    response = agent(user_query)

    print(json.dumps(response, indent=4))
```

# Conclusion

The single-agent system successfully routes queries, performs calculations, extracts keywords, handles unsupported requests, and returns structured JSON responses with proper error handling.
